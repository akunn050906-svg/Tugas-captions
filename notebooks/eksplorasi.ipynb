# Eksplorasi Dataset - OCR Plat Nomor Kendaraan
Notebook ini untuk EDA (exploratory data analysis) singkat sebelum training.
Jalankan setelah `src/download_dataset.py` dan `src/preprocess.py`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from pathlib import Path

PROC_DIR = Path('../data/processed')
labels = pd.read_csv(PROC_DIR / 'ocr' / 'labels.csv')
print('Jumlah data OCR:', len(labels))
labels.head()
```

```python
# Distribusi panjang teks plat
labels['label'].astype(str).str.len().hist(bins=15)
plt.title('Distribusi Panjang Teks Plat Nomor')
plt.xlabel('Jumlah Karakter')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Contoh visualisasi beberapa gambar plat
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, (_, row) in zip(axes, labels.sample(4).iterrows()):
    img_path = PROC_DIR / 'ocr' / 'images' / row['filename']
    img = cv2.imread(str(img_path))
    if img is not None:
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax.set_title(row['label'])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
